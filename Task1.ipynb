```python
# importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib
```

```python
# assesing the dataset
df=pd.read_csv("/content/Dataset .csv")
df
```

```python
# checking for the missing values
df.isnull().sum()
```

```python
# filling the missing values using most frequent ones
df['Cuisines'].fillna(df['Cuisines'].mode()[0],inplace=True)
```

```python
df.isnull().sum()
```

```python
# droping the target column from the dataset
X=df.drop('Aggregate rating',axis=1)
Y=df['Aggregate rating']
```

```python
# spliting the dataset into train and test
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error , r2_score
```

```python
# encoding the categorical data
X=pd.get_dummies(X)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
# scaling the feature
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
# training the model
model=RandomForestRegressor(n_estimators=100,random_state=42)
model.fit(X_train_scaled,Y_train)
```

```python
# predicting the output
Y_pred=model.predict(X_test_scaled)
```

```python
# evaluate the model
mse=mean_squared_error(Y_test,Y_pred)
rscore=r2_score(Y_test,Y_pred)

print(f"mean squared error : {mse}")
print(f"r2 score : {rscore}")
```

Mean Squared Error (MSE) of 0.03 indicates that model's predictions is very accurate and having less erorr

r2 score of 0.98 suggest that model is highly effective in explaining & predicting the target
